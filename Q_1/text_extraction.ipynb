```python
import os, re
```

```python
raw_dataset_path = "data/raw_data/"
```

```python
extracted_dataset_path = "data/extracted_data/"
```

```python
c = 0
for each_file in os.listdir(raw_dataset_path):
    current_file_path = os.path.join(raw_dataset_path, each_file)
    save_file_path = os.path.join(extracted_dataset_path, each_file)
    with open(current_file_path, "r") as file:
        temp = file.read()
    file.close()
    if c < 5:
        print("Current File: ", current_file_path.split('/')[-1])
        print("Original Text: ", temp)
    temp = temp.replace("\n", " ")
    title_text = re.search(r"<TITLE>(.*)</TITLE>", temp).group(1)
    body_text = re.search(r"<TEXT>(.*)</TEXT>", temp).group(1)
    
    final_text = title_text + " " + body_text
    if c < 5:
        print("Extracted Text: ", final_text)
        c += 1
    with open(save_file_path, "w") as file:
        file.write(final_text)
    file.close()
```
